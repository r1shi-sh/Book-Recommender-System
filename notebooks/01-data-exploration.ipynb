```python
import pandas as pd

# --- Configuration ---
RAW_DATA_DIR = "../data/raw/"

# --- Load Data ---
books = pd.read_csv(f"{RAW_DATA_DIR}Books.csv", sep=";", low_memory=False, on_bad_lines='skip', encoding='latin-1')
users = pd.read_csv(f"{RAW_DATA_DIR}Users.csv", sep=";", low_memory=False, on_bad_lines='skip', encoding='latin-1')
ratings = pd.read_csv(f"{RAW_DATA_DIR}Ratings.csv", sep=";", low_memory=False, on_bad_lines='skip', encoding='latin-1')

print("--- Books Info ---")
books.info()
print("\n--- Users Info ---")
users.info()
print("\n--- Ratings Info ---")
ratings.info()

print("\n--- First 5 Books ---")
display(books.head())

print("\n--- First 5 Users ---")
display(users.head())

print("\n--- First 5 Ratings ---")
display(ratings.head())
```
